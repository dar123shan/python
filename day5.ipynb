# **Visualizing Dummy Data**

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([3,6,13,28,55])
```

### Leanear Regration

```python
linear_model=LinearRegression()
linear_model.fit(x,y)
y_pred_linear=linear_model.predict(x)
```

## Polynomial Regressaion(degree 2)

```python
poly=PolynomialFeatures(degree=2)
x_poly=poly.fit_transform(x)
poly_model=LinearRegression()
poly_model.fit(x_poly,y)
y_pred_poly=poly_model.predict(x_poly)
```

## Plot

```python
plt.scatter(x,y,color='red',label='Actual')
plt.plot(x, y_pred_linear,label='Linear Fit',color='blue')
plt.plot(x, y_pred_poly,label='polymomial Fit(deg=2)',color='green')
plt.title("Linear vs Polynomial Fit(Dummy Date)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

## Real Dataset:salary prediction

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error,r2_score
```

## load dataset

```python
df=pd.read_csv("https://raw.githubusercontent.com/Naresh312143/mahalingpur-python-ml/refs/heads/main/Salary_Data.csv")
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

### Train_test split

```python
x=df[["Years of Experience"]]
y=df["Salary"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

#Linear Regation

## Train linear regration

```python
line_model=LinearRegression()
line_model.fit(x_train,y_train)
y_pred_line=line_model.predict(x_test)
```

# Matrix

```python
mse_line=mean_squared_error(y_test,y_pred_line)
rmse_line=np.sqrt(mse_line)
r2_line=r2_score(y_test,y_pred_line)

print("Linear Regration:")
print("MSE:",mse_line)
print("RMSE:",rmse_line)
print("R2 Score:",r2_line)
```

# polynomialRegration(Digree 2 and 4)

## Degree

```python
poly2=PolynomialFeatures(degree=2)
x_train_poly2=poly2.fit_transform(x_train)
x_test_poly2=poly2.transform(x_test)
poly2_model=LinearRegression()
poly2_model.fit(x_train_poly2,y_train)
y_pred_poly2=poly2_model.predict(x_test_poly2)
```

```python
x_train_poly2
```

```python
x_test_poly2
```

## Metrixs

```python
mse_poly2=mean_squared_error(y_test,y_pred_poly2)
rmse_poly2=np.sqrt(mse_poly2)
r2_poly2=r2_score(y_test,y_pred_poly2)
```

## Degree 4

```python
poly4=PolynomialFeatures(degree=4)
x_train_poly4=poly4.fit_transform(x_train)
x_test_poly4=poly4.transform(x_test)

poly4_model=LinearRegression()
poly4_model.fit(x_train_poly4,y_train)
y_pred_poly4=poly4_model.predict(x_test_poly4)
```

```python
x_train_poly4
```

```python
mse_poly4=mean_squared_error(y_test,y_pred_poly4)
rmse_poly4=np.sqrt(mse_poly4)
r2_poly4=r2_score(y_test,y_pred_poly4)

print("\nPolynomial Regression (Degree 2):")
print("MSE:", mse_poly2)
print("RMSE:", rmse_poly2)
print("R2 Score:", r2_poly2)

print("\nPolynomial Regression (Degree 4):")
print("MSE:", mse_poly4)
print("RMSE:", rmse_poly4)
print("R2 Score:", r2_poly4)
```

## Visualization of All Models

Generates smooth line for predictions

```python
X_range=np.linspace(x.min(),x.max(),100).reshape(-1,1)

#Create Dataframe with column name
X_range_df=pd.DataFrame(X_range,columns=['Years of Experience'])


#transform using polynomial features
X_range_poly2=poly2.transform(X_range_df)
X_range_poly4=poly4.transform(X_range_df)


#predict using models
y_range_line=line_model.predict(X_range_df)
y_range_poly2=poly2_model.predict(X_range_poly2)
y_range_poly4=poly4_model.predict(X_range_poly4)

#plot
plt.scatter(x,y,color='red',label='Actual Data')
plt.plot(X_range,y_range_line,label='Linear Mdel',color='blue')
plt.plot(X_range,y_range_poly2,label='Polynomial(deg=2)',color='green')
plt.plot(X_range,y_range_poly4,label='Polynomial(deg=4)',color='orange')
plt.title("Salary Prediction")
plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.legend()
plt.grid(True)
plt.show()
```

## Model Performance Comparison

```python
metrix_df=pd.DataFrame({
    'Model':['Linear Regration','Polynomial(deg=2)','Polynomial(deg=4)'],
    'MSE':[mse_line,mse_poly2,mse_poly4],
    'RMSE':[rmse_line,rmse_poly2,rmse_poly4],
    'R2 Score':[r2_line,r2_poly2,r2_poly4]
})

metrix_df.style.format({
    'MSE':'{:.2f}',
    'RMSE':'{:.2f}',
    'R2 Score':'{:.4f}'
})
```
